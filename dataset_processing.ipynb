```python
import pandas as pd
```

```python
df = pd.DataFrame()
for i in ['books_50+.csv','books_art.csv', 'books_art1.csv', 'books_bus.csv', 'books_bus1.csv','books_child.csv', 'books_child1.csv','books_6000.csv','books_foreign.csv', 'books_foreign1.csv','books_gifts.csv','books_nat.csv','books_gum.csv','books_gum1.csv','books_hobby.csv','books_it.csv','books_law.csv','books_med.csv','books_med1.csv','books_prom.csv', 'books_prom1.csv','books_psy_.csv','books_religion.csv','books_religion1.csv','books_sport.csv','books_study.csv']:
    temp=pd.read_csv(i)
    df = pd.concat([df,temp])
df
```

```python
df = df.drop_duplicates()

df
```

```python
pd.DataFrame(data=dict(df.isna().sum()), index=range(0,1))
```

```python
df = df[df['annotation'].notna()]
```

```python
df = df[df['annotation'].apply(lambda x: len(x.split()) >= 40)]
```
